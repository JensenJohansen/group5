```python
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(data['POPULATION SERVED'], data['LONGITUDE'], cmap=cmap, n_levels=60, shade=True);
```

```python
plt.figure(figsize=(10,15))
plt.title="Longitude and Latitude Location of the Waterpoints"
sns.jointplot(y=data.LATITUDE.values,x=data.LONGITUDE.values,size=10)
plt.ylabel("LONGITUDE",fontsize=12)
plt.xlabel("LATITUDE",fontsize=12)
plt.show()
sns.despine
```
